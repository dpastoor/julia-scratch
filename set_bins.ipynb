```julia
function set_bins(x::DataArray{Float64, 1}, left::Array{Float64, 1}, right::Array{Float64, 1})
    out = fill(0, length(x))
    for i in 1:length(x) 
        for j in 1:length(left) 
            if ( (x[i] >= left[j]) && (x[i] < right[j]))
            out[i] = j;
                break
             else 
               continue
             end
        end
    end
    return out
end
```
